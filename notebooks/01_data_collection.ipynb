```python
# Sprint 1: Data Collection Summary
import pandas as pd
import os
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

seasons = [2019, 2020, 2021, 2022, 2023, 2024]
raw_dir = Path("../data/real")

summary = []
for year in seasons:
    path = raw_dir / f"laps_{year}.csv"
    df = pd.read_csv(path)
    summary.append({
        "Year": year,
        "Races": df["RoundNumber"].nunique(),
        "Drivers": df["Driver"].nunique(),
        "Total Laps": len(df),
        "Compounds": df["Compound"].unique().tolist()
    })

summary_df = pd.DataFrame(summary)
summary_df
```

```python
# check missing values across all seasons
all_dfs = []
for year in seasons:
    df = pd.read_csv(raw_dir / f"laps_{year}.csv")
    df["Year"] = year
    all_dfs.append(df)

combined = pd.concat(all_dfs, ignore_index=True)

missing = combined.isnull().sum()
missing_pct = (missing / len(combined) * 100).round(2)

pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_pct
}).query("`Missing Count` > 0").sort_values("Missing %", ascending=False)
```

```python
# compound distribution across seasons
compound_dist = combined.groupby(["Year", "Compound"]).size().unstack(fill_value=0)
compound_dist
```
